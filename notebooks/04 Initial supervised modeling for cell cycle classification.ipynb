# Functions Section:

```python
def multi_auc(y, y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    fpr = dict()
    tpr = dict()
    auc_scores = []
    for i in range(0,3):
        fpr[i], tpr[i], thresh = roc_curve(y_bin[:,i], y_pred[:,i])
        auc_scores.append(roc_auc_score(y_bin[:,i],y_pred[:,i]))



#     averaged interpolated ROC: (Macro Average)
    fpr[3] = np.unique(np.concatenate([fpr[i] for i in range(0,3)]))
    tpr[3] = np.zeros_like(fpr[3])
    for i in range(0,3):
        tpr[3] += interp(fpr[3],fpr[i],tpr[i])/3
    
    auc_scores.append(auc(fpr[3],tpr[3]))
    
#     full combined ROC: (Micro Average)
    fpr[4], tpr[4], thresh = roc_curve(y_bin.ravel(), y_pred.ravel())
#     auc_scores[4] = auc(fpr[4], tpr[4])
    auc_scores.append(auc(fpr[4], tpr[4]))

    return auc_scores
```

```python
# AUC scoring functions for gridsearch

def G1_AUC(y,y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    return roc_auc_score(y_bin[:,0],y_pred[:,0])

def S_AUC(y,y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    return roc_auc_score(y_bin[:,1],y_pred[:,1])

def G2M_AUC(y,y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    return roc_auc_score(y_bin[:,2],y_pred[:,2])

def Micro_AUC(y,y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    fpr, tpr, thresh = roc_curve(y_bin.ravel(), y_pred.ravel())
    return auc(fpr, tpr)

def Macro_AUC(y,y_pred):
    y_bin = label_binarize(y, classes = [0,1,2])
    fpr = dict()
    tpr = dict()
    auc_scores = []
    for i in range(0,3):
        fpr[i], tpr[i], thresh = roc_curve(y_bin[:,i], y_pred[:,i])

#     averaged interpolated ROC:
    fpr[3] = np.unique(np.concatenate([fpr[i] for i in range(0,3)]))
    tpr[3] = np.zeros_like(fpr[3])
    for i in range(0,3):
        tpr[3] += interp(fpr[3],fpr[i],tpr[i])/3
    
    return auc(fpr[3],tpr[3])
```

```python
def multi_roc(y, y_pred,label):
    y_bin = label_binarize(y, classes = [0,1,2])
    fpr = dict()
    tpr = dict()
    auc_scores = dict()
    for i in range(0,3):
        fpr[i], tpr[i], thresh = roc_curve(y_bin[:,i], y_pred[:,i])
        auc_scores[i] = roc_auc_score(y_bin[:,i],y_pred[:,i])

#     averaged interpolated ROC: (Macro Average)
    fpr[3] = np.unique(np.concatenate([fpr[i] for i in range(0,3)]))
    tpr[3] = np.zeros_like(fpr[3])
    for i in range(0,3):
        tpr[3] += interp(fpr[3],fpr[i],tpr[i])/3
    
    auc_scores[3] = auc(fpr[3],tpr[3])
    
#     full combined ROC: (Micro Average)
    fpr[4], tpr[4], thresh = roc_curve(y_bin.ravel(), y_pred.ravel())
    auc_scores[4] = auc(fpr[4], tpr[4])

    classes = ['G1','S','G2/M','Averaged','Combined']
    colors = ['aqua','darkorange','cornflowerblue','navy','deeppink']
    weights = [1,1,1,3,3]
    styles =['-','-','-','-.','-.']
    plt.figure(figsize=(12,8))
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1)
    for i in range(0,5):
        plt.plot(fpr[i],tpr[i],label=classes[i]+', AUC= %0.2f' % auc_scores[i],color=colors[i],
                linestyle=styles[i],linewidth=weights[i])
    
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title(label +' ROC curve')
    plt.legend(loc='best')
```

# Cross-validation to find best model and parameters

I will do a nested cross validation to tune the parameters of the three models and see which one does best.  Because I have a fairly small dataset, rather than the traditional folding, the nested cross validation will go as 10 independent train-test splits and then for each each, run a gridsearchCV on the training data. The best 4 models will be kept and the best model for each split will be evaluated by multi-class AUC.

```python
# Time Consuming!!
print datetime.datetime.now()

CV_comp = pd.DataFrame()
AUC_comp = pd.DataFrame(columns=['G1','S','G2/M','Averaged','Combined','Model'])

params_rf = {"n_estimators":[100,500],
             "max_depth": [3, 5],
             "max_features": ["sqrt", None]}

params_gbc = {"n_estimators":[100,500],
              "max_depth": [3, 5],
              "max_features": ["sqrt", None]}

params_xgb = {'max_depth':[4,6],
             'learning_rate':[0.05,0.1],
             'n_estimators':[70,140]}

rf = RandomForestClassifier()
gbc = GradientBoostingClassifier()
xgb = XGBClassifier(objective= 'multi:softprob', silent=1, num_class=3,
                                               eval_metric='merror', nthread=4, scale_pos_weight=1)

grid_search_rf = GridSearchCV(rf, param_grid=params_rf, scoring='accuracy')
grid_search_gbc = GridSearchCV(gbc, param_grid=params_gbc, scoring = 'accuracy')
grid_search_xgb= GridSearchCV(xgb, param_grid = params_xgb, scoring = 'accuracy')

print str(datetime.datetime.now()),'Starting...'

for i in range(1,12):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, stratify=y)
    print str(datetime.datetime.now()),'Training ',i,'th RF'
    grid_search_rf.fit(X_train,y_train)
    print str(datetime.datetime.now()),'Training ',i,'th GBC'
    grid_search_gbc.fit(X_train,y_train)
    print str(datetime.datetime.now()),'Training ',i,'th XGB'
    grid_search_xgb.fit(X_train,y_train)
    
    temp = pd.DataFrame(grid_search_rf.cv_results_).sort_values(['rank_test_score']).iloc[:4,:7]
    temp['Model']='RF'
    CV_comp=CV_comp.append(temp,ignore_index=True)
    temp = pd.DataFrame(grid_search_gbc.cv_results_).sort_values(['rank_test_score']).iloc[:4,:7]
    temp['Model']='GBC'
    CV_comp=CV_comp.append(temp,ignore_index=True)
    temp = pd.DataFrame(grid_search_xgb.cv_results_).sort_values(['rank_test_score']).iloc[:4,:7]
    temp['Model']='XGB'
    CV_comp=CV_comp.append(temp,ignore_index=True)

    # predict on the test data:
    y_pred_rf = grid_search_rf.predict_proba(X_test)
    y_pred_gbc = grid_search_gbc.predict_proba(X_test)
    y_pred_xgb = grid_search_xgb.predict_proba(X_test)
    auc_temp = multi_auc(y_test,y_pred_rf)
    auc_temp.append('RF')
    AUC_comp = AUC_comp.append(pd.Series(auc_temp,index=auc_columns),ignore_index=True)
    auc_temp = multi_auc(y_test,y_pred_gbc)
    auc_temp.append('GBC')
    AUC_comp = AUC_comp.append(pd.Series(auc_temp,index=auc_columns),ignore_index=True)
    auc_temp = multi_auc(y_test,y_pred_xgb)
    auc_temp.append('XGB')
    AUC_comp = AUC_comp.append(pd.Series(auc_temp,index=auc_columns),ignore_index=True)
```
