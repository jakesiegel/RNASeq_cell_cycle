```python
%matplotlib inline
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import datetime
import sys
sys.path.append("..")
import src as RNAseq #RNAseq helper functions
```

```python
data_src = '../data/processed/KarlssonTPM.csv'
df = pd.read_csv(data_src, sep = '\t')
df = df.set_index(list(df.columns)[0])
df = df.drop(['bp'],axis=0)
df.Stage = pd.factorize(df.Stage)[0]
```

# Initial test autoencoder to get a feel for how it works:

```python
# X_train, X_test, y_train, y_test = train_test_split(X_in, y, test_size=.25, stratify=y)

X_scaled = pd.DataFrame(StandardScaler().fit_transform(X_in))
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=.25, stratify=y)
```

```python
# I'll start from a simple autoencoder example, merging it with Lin et al, NAR 2017 doi.org/10.1093/nar/gkx681
# The Lin paper uses a denoising autoencoder with nodes of 796 and 100 in the two hidden layers, respectfully.
# It should go from input_dim->796->100->796->input_dim
# let me set this up without the corruption function, for now.

# Training Parameters
learning_rate = 0.1
num_steps = 10000
batch_size = 12

display_step = 1000

# Network Parameters
num_hidden_1 = 796 # 1st layer num features from Lin et al.
num_hidden_2 = 100 # 2nd layer num features from Lin et al. (the latent dim)
num_input = X_train.shape[1] # input number of genes

# tf Graph input
X = tf.placeholder("float", [None, num_input])

weights = {
    'encoder_h1': tf.Variable(tf.random_normal([num_input, num_hidden_1])),
    'encoder_h2': tf.Variable(tf.random_normal([num_hidden_1, num_hidden_2])),
    'decoder_h1': tf.Variable(tf.random_normal([num_hidden_2, num_hidden_1])),
    'decoder_h2': tf.Variable(tf.random_normal([num_hidden_1, num_input])),
}
biases = {
    'encoder_b1': tf.Variable(tf.random_normal([num_hidden_1])),
    'encoder_b2': tf.Variable(tf.random_normal([num_hidden_2])),
    'decoder_b1': tf.Variable(tf.random_normal([num_hidden_1])),
    'decoder_b2': tf.Variable(tf.random_normal([num_input])),
}

# Building the encoder
def encoder(x):
    # Encoder Hidden layer with tanh activation #1
    layer_1 = tf.tanh(tf.add(tf.matmul(x, weights['encoder_h1']),
                                   biases['encoder_b1']))
    # Encoder Hidden layer with tanh activation #2
    layer_2 = tf.tanh(tf.add(tf.matmul(layer_1, weights['encoder_h2']),
                                   biases['encoder_b2']))
    return layer_2

# these should also be be equivalent to something like:
# layer_1 = tf.layers.dense(x, ...[weights and biases]..,activation= tf.tanh)

# Build the decoder (symmetric):
def decoder(x):
    # Decoder Hidden layer 1 with tanh activation 1
    layer_1 = tf.tanh(tf.add(tf.matmul(x, weights['decoder_h1']),
                            biases['decoder_b1']))
    # Decoder Hidden layer 2 with tanh activation 2
    layer_2 = tf.tanh(tf.add(tf.matmul(layer_1,weights['decoder_h2']),
                            biases['decoder_b2']))
    return layer_2

# Construct model
encoder_op = encoder(X)
decoder_op = decoder(encoder_op)

# now I need to set up the prediction and the output prediction and the input to train the data:
y_pred = decoder_op
y_true = X

# Now I define an optimizer and loss function to minimize: as a first pass, let's do MSE with gradient descent
loss = tf.reduce_mean(tf.pow(tf.subtract(y_true, y_pred), 2))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
# TF has a bunch of built in optimizers, eventually I will probably switch this to an adaptive learning rate
# optimizer like RMSprop or Adadelta

# initialize
init = tf.global_variables_initializer()

# training:
with tf.Session() as sess:

    # Run the initializer
    sess.run(init)

    # Training
    for i in range(1, num_steps+1):
        # Prepare Data:
        batch_x = X_train.iloc[np.random.choice(X_train.shape[0],batch_size),:]

        # Run optimization op (backprop) and cost op (to get loss value)
        _, l = sess.run([optimizer, loss], feed_dict={X: batch_x.values})
        # Display logs per step
        if i % display_step == 0 or i == 1:
            print('Step %i: Minibatch Loss: %f' % (i, l))
            
    # Testing:
    test_reconstruction = sess.run(decoder_op, feed_dict={X:X_test.values})
    test_reduction = sess.run(encoder_op, feed_dict={X:X_test.values})
```

```python
test_recon = pd.DataFrame(test_reconstruction)
test_encoded = pd.DataFrame(test_reduction)
test_recon.head()
```
