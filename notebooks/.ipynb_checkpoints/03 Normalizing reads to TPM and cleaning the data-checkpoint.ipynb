```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in read counts and put in cell cycle stage, as done in notebook 01

data_src = '../data/raw/counts.txt'
df = pd.read_csv(data_src, sep = '\t')

info_src = '../data/raw/E-MTAB-6142.sdrf.txt'
df_info = pd.read_csv(info_src, sep = '\t')

df_info = df_info.iloc[:,-2:].drop_duplicates().transpose().reset_index().drop(columns=['index'],axis=1)
df_info.columns = df_info.iloc[1]
df_info = df_info.drop(1)
df = df.set_index('Gene_ID')
df = pd.concat([df_info.rename(index={0: 'Stage'}),df]).transpose()
df
```

Before merging the transcript lengths, let's eliminate the spike-in controls

```python
cols = [c for c in df.columns if c[0:4]!='ERCC']
df = df[cols]
```

Load in the transcript lengths and concatenate

```python
# transcript_lengths.to_csv('../data/interim/transcipt_lengths.csv', encoding='utf-8', index=False)
lengths_src = '../data/interim/transcipt_lengths.csv'
df_lengths = pd.read_csv(lengths_src)
df_lengths.head()
```

```python
# Concatenate transcript lengths
df_lengths = df_lengths.set_index('gene_id').transpose()
df = pd.concat([df,df_lengths])

# remove transcripts that do not appear in the counts.txt file
observed = df.isnull().sum(axis=0)<2
df = df[observed.index[observed]]
df
```

### Reads per transcript

Before normalizing to TPM, I want to take a quick look at the distribution of reads in the transcripts to determine if I should eliminate any transcripts that have no reads at all.

```python
df.shape
```

```python
# Looking at the number of transcipts with less than or equal to 100 reads, by read count
low_reads = pd.DataFrame()
for i in range(0,100):
    lowcounts = df.iloc[:,:-1].sum(axis=0)==i
    temp = pd.Series([i,lowcounts.sum()],index=['reads','number_transcripts'])
    low_reads = low_reads.append(temp,ignore_index=True)
    
low_reads.plot(x='reads',y='number_transcripts')
```

```python
# It looks like the number of transcripts flattens to zero at low read counts; let's zoom in
low_reads.iloc[:45].plot(x='reads', y='number_transcripts')
```

```python
# Looking at the number of transcripts with at least powers of 10 reads
binned_reads = pd.DataFrame()
lowcounts = df.iloc[:,:-1].sum(axis=0)!=0
temp = pd.Series([0,lowcounts.sum()],index=['min_reads','number_transcripts'])
binned_reads = binned_reads.append(temp,ignore_index=True)
for i in range(0,7):
    lowcounts = df.iloc[:,:-1].sum(axis=0)>=pow(10,i)
    temp = pd.Series([pow(10,i),lowcounts.sum()],index=['min_reads','number_transcripts'])
    binned_reads = binned_reads.append(temp,ignore_index=True)

# display the table
binned_reads
```

Let's see if some magically happened to drop out only for certain cell cycle stages.

```python
# First see if the distribution of observed transcripts by cell number changes

stages = df['Stage'].dropna().unique()
cols = np.append('cell',stages)
zero_reads = pd.DataFrame()
for i in range(1,33):
    temp = pd.Series(i)
    for stage in stages:
        zero_counts = df.loc[df['Stage']==stage].astype(bool).sum(axis=0)==i
        temp = temp.append(pd.Series(zero_counts.sum()),ignore_index=True)
    zero_reads = zero_reads.append(temp, ignore_index=True)
    
zero_reads.columns=cols    
zero_reads.head()
zero_reads.plot(x='cell')
```

```python
#  To see if there are some transcripts that are only observed in a given cell cycle
cell_cycle = df.groupby('Stage').apply(lambda x: x.astype(bool).sum())
cell_cycle
```

```python
keepcols = []

for column in cell_cycle:
    keepcol = False
    if abs(cell_cycle[column]['G1']-cell_cycle[column]['G2/M']) > 10:
        keepcol = True
    if abs(cell_cycle[column]['G1']-cell_cycle[column]['S']) > 10:
        keepcol = True
    if abs(cell_cycle[column]['G2/M']-cell_cycle[column]['S']) > 10:
        keepcol = True
    if keepcol:
        keepcols.append(column)
        
cell_cycle_specific = cell_cycle[keepcols]
cell_cycle_specific
```

# TPM Normalization

The TPM calculation is (reads/transcript length in kb)/ 1e-6*sum(reads / transcript length in kb)

```python
df
```

```python
tpm = df.iloc[:,:-1].copy()
tpm = 1e3*tpm.div(tpm.loc['bp'])
tpm['sum']=1e-6*tpm.sum(axis=1)
tpm = tpm.apply(lambda x: x/tpm['sum'])
tpm = tpm.drop(['sum'],axis=1)

tpm['Stage']=df['Stage']
tpm
```

```python
mean_tpm_by_cell = pd.DataFrame()
tpm2 = tpm.iloc[:,:-1]
for column in tpm2:
    temp= pd.Series([tpm2[column].astype(bool).sum(axis=0),tpm2[column].replace(0,np.NaN).mean(axis=0)],
                    index=['n_cells_where_transcript_detected','mean_tpm_by_transcript'])
    mean_tpm_by_cell = mean_tpm_by_cell.append(temp,ignore_index=True)
    
mean_tpm_by_cell.plot.scatter(x='n_cells_where_transcript_detected',y='mean_tpm_by_transcript',figsize=[12,8], 
                                 logy=True,alpha=0.1)
```

```python
mean_tpm_by_cell = pd.DataFrame()
tpm2 = tpm.iloc[:,:-1]
for column in tpm2:
    temp= pd.Series([tpm2[column].astype(bool).sum(axis=0),tpm2[column].mean(axis=0)],
                    index=['n_cells_where_transcript_detected','mean_tpm_by_transcript'])
    mean_tpm_by_cell = mean_tpm_by_cell.append(temp,ignore_index=True)
    
mean_tpm_by_cell.plot.scatter(x='n_cells_where_transcript_detected',y='mean_tpm_by_transcript',figsize=[12,8], 
                                 logy=True,alpha=0.1)
```

While many transcripts are not observed in the majority of most cells, I think it is likely that the cut-off of detection in at least 15 cells or more is too strict.  Looking at the mean TPM vs number of cells the transcript was detected, I believe 7 or 8 is the probably correct cut-off, although it depends on the purpose.  Bare minimum, the ~38,000 transcripts detected in only 1 cell should be eliminated.
